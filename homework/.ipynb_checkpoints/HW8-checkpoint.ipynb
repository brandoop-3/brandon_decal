```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

###### PROBLEM 1############
###### PART 1 #######
def make_sine_wave(x,A,w):
    return A*np.sin(w*x)

####### PART 2 ###########
x_arr = np.linspace(0, 2*np.pi, 1000) # x array

###### PART 3 ##############
lines = ["-", "--", "-", ":", "-."] #list of different line styles
line_colors = ['red', 'orange', 'green', 'purple', 'blue']
line_widths = [0.5, 1, 1.5, 2, 2.5]
amp_arr = ([0.5, 1, 1.5, 2, 2.5]) #amplitude values
freq_arr = ([1, 2, 3, 4, 5]) #frequency values

######### PART 4 #############
plt.figure(figsize = (10,6))
```

```python
######### PART 4 #############
plt.figure(figsize = (10,6))

############ PART 5 #############
params = list(zip(amp_arr, freq_arr, lines, line_colors, line_widths)) #use of the zip() function

#run through all tuples (frequencies, amplitudes, and line styles)
for A, w, line_style, line_colors, line_widths in params:
    y = make_sine_wave(x_arr,A,w) #calling function from above
    plt.plot(x_arr, y, ls = line_style, color = line_colors, linewidth = line_widths, label =  f'A = {A}, w = {w}' )

#plotting 
plt.title("Different Sine Waves for Different Amplitudes and Frequencies")
plt.xlabel("x-values")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show
```

```python
##### PART 3 ############
mass_array = np.array(df_stars['Mass (M)']) #creates mass column into numpy array
#print(mass_array) 
avg_mass = np.mean(mass_array) #takes the average of the array
print(avg_mass)

temp_array = np.array(df_stars['Temperature (K)']) #creates temp column into numpy array
#print(temp_array)
avg_temp = np.mean(temp_array) #takes the average of the temp array
print(avg_temp)

radius_array = np.array(df_stars['Radius (R)']) # creates radius ccolumn into numpy array
#print(radius_array)
max_radius = np.max(radius_array) #takes the maximum value from the array
print(max_radius)

ST_array  = np.array(df_stars['Spectral_Type']) #creates ST array into numpy array
#print(ST_array) 

#the lines below
    # - interates to first string
    # - iterates through every character to find 'M' which will always be first character anyways
    # - if 'M' is found, add 1 to sum

sum = 0 
for s in ST_array:
    for char in s: 
        if char == 'M':
            sum += 1

print(sum) 
```
